```python
# square
n=int(input())
for i in range(n):   # Rows
    for j in range(n):   # Columns
        print('*', end="")
    print()
```

```python
# Increasing triangle
n=int(input())
for i in range(n):   # Rows from 0 to n(Rows)
    for j in range(i+1):   # Here we are increasing columns with R+1 
        print('*', end="")
    print()
```

```python
# Decreasing triangle
n=int(input())
for i in range(n):   
    for j in range(i,n):  # Here we are decreasing columns with R-1 
        print('*', end="")
    print()
```

```python
# Decreasing triangle
n=int(input())
for i in range(n):   
    for j in range(n-i):  # Here as row increases columns decreses by row's no
        print('*', end="")
    print()
```

```python
# Right sided triangle
n=int(input())
for i in range(n):   
    for j in range(i,n):   # Decreasing triangle
        print(' ', end="")
    for j in range(i+1):   # Increasing triangle
        print('*', end="")
    print()
```

```python
# Inverted Right sided triangle
n=int(input())
for i in range(n):   
    for j in range(i+1):   # Increasing triangle
        print(' ', end="")
    for j in range(i,n):   # Decreasing triangle
        print('*', end="")
    print()
```

```python
# Hill Pattern
n=int(input())
for i in range(n):  
    for j in range(i,n):   # Decreasing triangle
        print(' ', end="")
    for j in range(i):   # Increasing triangle, Here only 'i' is for to print one less column
        print('*', end="")
    for j in range(i+1):   # Increasing triangle
        print('*', end="")
    print()
```

```python
# Reverse Hill Pattern
n=int(input())
for i in range(n):  
    for j in range(i+1):   # Increasing triangle
        print(' ', end="")
    for j in range(i,n-1):   # Decreasing triangle
        print('*', end="")
    for j in range(i,n):   # Decreasing triangle
        print('*', end="")
    print()
```

```python
# Diamond pattern :- Hill Pattern & Reverse Hill Pattern
n=int(input())

# Hill
for i in range(n-1):  
    for j in range(i,n):   # Decreasing triangle
        print(' ', end="")
    for j in range(i):   # Increasing triangle
        print('*', end="")
    for j in range(i+1):   # Increasing triangle
        print('*', end="")
    print()
    
# Reverse Hill
for i in range(n):  
    for j in range(i+1):   # Increasing triangle
        print(' ', end="")
    for j in range(i,n-1):   # Decreasing triangle
        print('*', end="")
    for j in range(i,n):   # Decreasing triangle
        print('*', end="")
    print()
```

```python
# Half upper Hallow diamond
r=4
for i in range(r):
    print(' '*(r-i-1)+'*',end="")  # space before left'*'
    if i>=1:     
        print(' '*(2*i-1)+'*',end="")  # There are odd no.of spaces between '*'
    print()
```

```python
# https://www.youtube.com/watch?v=8TtldUC44aU&list=PLd3UqWTnYXOmHGIurNDgStXNOsxjItRAz&index=17
# Half lower Hallow diamond
n=4
for i in range(n):
    print(' '*i+'*',end="")  
    if i!=n-1:     
        print(' '*(2*n-2*i-3)+'*',end="")  # There are odd no.of spaces between '*'
    print()
```

```python
# Butterfly 
r = 3

# Upper Triangles
for i in range(1,r+1):
    print("*"*i, end="")   # leftmost upper right-angled triangle
    print(" "*(r-i)*2, end="")   # spaces in between
    print("*"*i)                # rightmost upper left-angled triangle.

# Lower Triangles
for i in range(r,0,-1):
    print("*"*i, end="")
    print(" "*(r-i)*2, end="")
    print("*"*i)    
```
