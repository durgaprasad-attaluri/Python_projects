Whenever there is an error in a program, exceptions are raised.

Exception handling in Python is achieved using the try and except blocks. The try block is used to enclose the code that may raise an exception, while the except block is used to handle the exception that may occur. If an exception is raised in the try block, the program jumps to the except block to handle it, instead of halting the program altogether.

**Syntax**
<pre>
try:
    # code that may raise an exception
    
except:
    # code to handle the exception</pre>

```python
try:
    a=b
except Exception as e:
    print(e)
```

```python
#ArithmeticError ->during numeric calculation
#some are 'ZeroDivisionError;,'OverflowError','FloatingPointError','UnderflowError','TimeoutError'
def perform_arithmetic_operation(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Division by zero!")
    except OverflowError:
        print("Error: Arithmetic operation resulted in an overflow!")
    except FloatingPointError:
        print("Error: Floating-point operation failed!")
    except ArithmeticError:
        print("Error: An arithmetic error occurred!")
    except Exception:
        print("Error: Something went wrong!")

# Example usage
perform_arithmetic_operation(10, 0)  # ZeroDivisionError
perform_arithmetic_operation(1e1000, 1e-1000)  # OverflowError
perform_arithmetic_operation(0.1, 0.0)  # FloatingPointError
perform_arithmetic_operation(10, '2')  # TypeError (not an ArithmeticError)
```

```python
#keyError
def get_dictionary_value(dictionary, key):
    try:
        return dictionary[key] 
    except KeyError:
        raise KeyError("Key '" + str(key) + "' not found in the dictionary!")

# Example usage
my_dict = {"a": 1, "b": 2, "c": 3}

try:
    value = get_dictionary_value(my_dict, "d")
    print("Value:", value)
except KeyError as e:
    print("Error:", str(e))
```

```python
#TypeError
def add_numbers(a, b):
    try:
        result = a + b
        return result
    except TypeError:
        raise TypeError("Unsupported operand types: " + str(type(a)) + " and " + str(type(b)))

# Example usage
try:
    sum = add_numbers(5, "10")
    print("Sum:", sum)
except TypeError as e:
    print("Error:", str(e))
```

```python
#AttributeError
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello, my name is", self.name)

# Example usage
person = Person("Alice")

try:
    person.age  # Accessing a non-existent attribute
except AttributeError as e:
    print("Error:", str(e))
```

```python
#creating a custom exception
def custom_exception(message):
    return Exception(message)

def divide(a, b):
    if b == 0:
        raise custom_exception("Division by zero is not allowed!")
    return a / b

try:
    result = divide(10, 0)
except Exception as e:
    print("Error:", str(e))
```

```python
# define Python user-defined exceptions
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass

number = 18

try:
    age = int(input("Enter your age: "))
    if age < number:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print(f"Invalid Age. You can vote after {number-age} years.")
```
